Determine a way to use this vectorization to suggest the closest items to the entry in the database

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def vectorizer_search(query : str,
                      vectors=vectors,
                      vectorizer=vectorizer, top = 5) -> list:
    #TODO
    vec_query = vectorize_query(query,vectorizer)
    cos_sim = cosine_similarity(vec_query,vectors)[0]

    #créer une liste de tuple (index du doc, similarité avec la requête)
    sim = []
    for i in range(len(cos_sim)):
      if cos_sim[i] > 0:
        sim.append((i,cos_sim[i]))
    sim.sort(key = lambda x : x[1], reverse = True)

    #retrouver les textes du top de similarité
    top_sim = sim[:top]
    res = []
    for elem in top_sim:
      text = posts['cleaned_body'][elem[0]]
      res.append(text)

    return res
```

```python
entry = 'what is stochastic gradient descent ?'
```

```python
vectorizer_search(entry, top = 5)
```
